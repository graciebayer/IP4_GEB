```python
sns.set(rc={'figure.figsize':(12,10)})
heat_map=df[['Sales_in_thousands', '__year_resale_value', 'Price_in_thousands', 'Engine_size',
                   'Horsepower', 'Wheelbase', 'Width', 'Length', 'Curb_weight', 'Fuel_capacity',
                   'Fuel_efficiency', 'Power_perf_factor']].corr(method='pearson')
cols=['Sales_in_thousands', '__year_resale_value', 'Price_in_thousands', 'Engine_size',
                   'Horsepower', 'Wheelbase', 'Width', 'Length', 'Curb_weight', 'Fuel_capacity',
                   'Fuel_efficiency', 'Power_perf_factor']
axis=sns.heatmap(heat_map, annot=True, xticklabels=cols, yticklabels=cols)
```

```python
top_50=df.sort_values('Manufacturer', ascending=False).head(50)
top_50
```

```python
scatter_plot=sns.scatterplot(x='Manufacturer', y='Sales_in_thousands', data=top_50, color='red')
```

```python
plt.scatter(df['Sales_in_thousands'],df['Manufacturer'])
plt.title('Car Sales per Manufacturer')
plt.xlabel('Sales (in Thousands)')
plt.ylabel('Manufacturer')
plt.show()
```

```python
top_20=df.sort_values('Model',ascending=False).head(20)
top_20
plt.bar(top_20['Manufacturer'],top_20['Model'])
plt.ylabel('Model')
plt.xlabel('Manufacturer')
plt.title('Model based on Manufacturer')
plt.show()
```

```python
pair_plot=sns.pairplot(df)
```

```python
explode = (0.1,0)  
fig1, ax1 = plt.subplots(figsize=(12,7))
ax1.pie(df['Vehicle_type'].value_counts(), explode=explode,labels=['Passenger','Car'], autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
plt.tight_layout()
plt.legend()
plt.show()
```
